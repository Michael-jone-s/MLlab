```python
#importing libraries
import datetime as dt
import pandas as pd
import numpy as np
```

```python
#creating dataframe
dates= pd.Series(pd.date_range('2019-12-01', freq='31H15min',periods=5))
df= pd.DataFrame(dict(date=dates))
df.head()
```

```python
#year feature
df['year']= df.date.dt.year
df.head()
```

```python
#Month feature
df['Month']= df.date.dt.month
df.head()
```

```python
#day feature
df['Day']= df.date.dt.day
df.head()
```

```python
#Hour feature
df['Hour']= df.date.dt.hour
df.head()
```

```python
#minute feature
df['Minute']= df.date.dt.minute
df.head()
```

```python
#Day of the year feature
df['day_of_year']= df.date.dt.dayofyear
df.head()
```

```python
#week day feature
df['day_of_week']= df.date.dt.dayofweek
df.head()
```

```python
#name of the day feature
df['day_name']= df.date.dt.day_name()
df.head()
```

```python
#quarter of the year feature
df['quarter_of_year']= df.date.dt.quarter
df.head()
```
